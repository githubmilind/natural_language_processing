<b>Assignment - Text processing using NLTK</b>

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk import word_tokenize, pos_tag
import string
```

<b>Get the list of English stop words</b>

```python
stop_words = set(stopwords.words('english'))
```

<b>Read the Text File</b>

```python
with open('news article.txt') as f:
    content = f.read()
```

<b>Split the text in to Sentences</b>

```python
sentences = nltk.sent_tokenize(content)
```

<b>Loop over sentences and process each sentence</b>

```python
for sentence in sentences:
    
    # print sentence
    print(sentence)
    
    # tokensize the sentence in to array of words
    tokenized_text = nltk.word_tokenize(sentence)
    
    # preprocessing the words: Remove the stop words and punctuations
    tokenized_text_no_stop_words = [word for word in tokenized_text 
                                    if not ((word in stopwords.words()) or (word in string.punctuation))]
    
    # Tag each token with POS - part of speech 
    pos_tagged = nltk.pos_tag(tokenized_text_no_stop_words)
    print("\nPOS:\n", pos_tagged)
    
    # Find named entites
    named_entites = nltk.ne_chunk(pos_tagged)
    print("\nNAMED ENTITIES:\n", named_entites)
    
    # print dashed line
    print("---------------------------------------------------------------------------------------------------")
```
