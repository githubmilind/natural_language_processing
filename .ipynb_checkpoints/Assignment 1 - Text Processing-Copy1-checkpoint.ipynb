```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
from nltk import word_tokenize, pos_tag
import string
stop_words = set(stopwords.words('english'))
```

<b>Read Text File</b>

```python

with open('news article.txt') as f:
    for line in f:
        text_line = line.strip()
        #text_line = "NASA awarded Elon Muskâ€™s SpaceX a $2.9 billion contract to build the lunar lander."
        print(text_line)
        
        tokenized = word_tokenize(text_line)
        
        for i in tokenized:
            print("0>> ", i)
            workToken = word_tokenize(i)
            print("1>> ", workToken)
            
            workToken = [w for w in workToken if not w in stop_words]
            print("--> ", workToken)
            
            tagged = nltk.pos_tag(workToken)
            print("==> ")
            for t in tagged:
                print("a> ", t)
            
            ne_tree = nltk.ne_chunk(tagged)
            print(">>> ", ne_tree)
 
        break
```

```python
with open('news article.txt') as f:
    content = f.read()
tokens = word_tokenize(content)
tags = nltk.pos_tag(tokens)
tags
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(tags)
print(cs)
```

```python
ne_tree = nltk.ne_chunk(pos_tag(word_tokenize(content)))
print(ne_tree)
```

```python


sent_text = nltk.sent_tokenize(content) # this gives us a list of sentences
# now loop over each sentence and tokenize it separately
for sentence in sent_text:
    tokenized_text = nltk.word_tokenize(sentence)
    
    tokens_without_sw = [word for word in tokenized_text if not ((word in stopwords.words()) or (word in string.punctuation))]
    
    print(tokens_without_sw)
    tagged = nltk.pos_tag(tokens_without_sw)
    print(tagged)
    ne = nltk.ne_chunk(tagged)
    print(ne)
```
